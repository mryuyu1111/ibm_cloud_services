```python
import pandas as pd
import numpy as np
from ibm_watson import DiscoveryV1
from ibm_watson import ApiException
import json
```

```python
with open('ibm-credentials.env', 'r') as f:
    text = f.read().splitlines()
    api_key = text[0].split('=')[1]
    url = text[1].split('=')[1]

# print(api_key)
# print(url)
```

```python
discovery = DiscoveryV1(version='2019-03-25', iam_apikey=api_key, url=url)
```

```python
# query(self, environment_id, collection_id, filter=None, query=None, natural_language_query=None, 
# passages=None, aggregation=None, count=None, return_fields=None, offset=None, sort=None, highlight=None, 
# passages_fields=None, passages_count=None, passages_characters=None, deduplicate=None, deduplicate_field=None, 
# collection_ids=None, similar=None, similar_document_ids=None, similar_fields=None, bias=None, logging_opt_out=None, **kwargs)

response = discovery.query('system', 'news-en', query='IBM')
```

```python
result = response.get_result()

# number of returned results
print(result['matching_results'])
```
